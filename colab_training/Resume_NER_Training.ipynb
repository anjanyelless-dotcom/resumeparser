## Step 3: Upload Data Files

**IMPORTANT**: Upload these 3 files:
- `labelfiledata.json`
- `labelfiledata1.json`
- `labelfiledata2.json`

Click the folder icon on the left → Upload files

````python
import os

# Check if files exist
required_files = [
    "labelfiledata.json",
    "labelfiledata1.json",
    "labelfiledata2.json"
]

print("📋 Checking for data files...\n")
all_found = True
for filename in required_files:
    if os.path.exists(filename):
        size_mb = os.path.getsize(filename) / (1024 * 1024)
        print(f"✅ {filename} ({size_mb:.1f} MB)")
    else:
        print(f"❌ {filename} - NOT FOUND")
        all_found = False

if all_found:
    print("\n✅ All files found! Ready to train.")
else:
    print("\n❌ Please upload missing files before continuing.")
````

## Step 4: Upload Training Script

````python
# Upload colab_train.py file
from google.colab import files

print("📤 Upload colab_train.py file...")
uploaded = files.upload()

if 'colab_train.py' in uploaded:
    print("✅ Training script uploaded!")
else:
    print("❌ Please upload colab_train.py")
````

## Step 6: Download Trained Model

After training completes, download the model folder.

````python
import shutil

# Zip the model folder
print("📦 Creating ZIP file...")
shutil.make_archive('trained_model', 'zip', 'trained_model')
print("✅ Model zipped!")

# Download
print("\n📥 Downloading model...")
files.download('trained_model.zip')
print("✅ Download complete!")
````
